```python
import pandas as pd
import json
import urllib.parse

# Load JSON file
with open("instagram scraping_original_23032025.json", "r", encoding="utf-8") as file:
    data = json.load(file)  # Load JSON as a Python object
```

```python
# Convert JSON to DataFrame
df = pd.json_normalize(data, sep="_")  # `sep` replaces nested keys with "parent_child"
df.fillna("", inplace=True)  # Replace NaN with an empty string
df
```

```python
df = df.applymap(lambda x: urllib.parse.unquote(x) if isinstance(x, str) else x)
df
```

```python
df = df.drop(df.columns[3:9], axis=1)
df
```

```python
# df.to_csv("output.csv", index=False)  # Set index=False to avoid writing row numbers

# print("CSV file saved successfully.")
```

```python
df= df.drop(df.columns[1:2], axis=1)
df
```

```python
columns = [
    "inputUrl", "id", "type", "shortCode", "caption", "hashtags", "url",
    "commentsCount", "likesCount", "ownerFullName", "ownerUsername",
    "videoDuration", "isSponsored"
]

# Create an empty DataFrame with the specified column names
ds = pd.DataFrame(columns=columns)

# Display the empty DataFrame
print(ds)
```

```python
df
```

```python
for i in range(123):
    # print(df.iloc[0,i])
    dummy = pd.json_normalize(df.iloc[i,1])
    ds = pd.concat([ds, dummy], ignore_index=True)
    dummy = None

ds
```
